```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams


# Load data
X_comin = np.load("../data/sample/comin_dataset.npy")
y_comin = np.load("../data/sample/comin_labels.npy")
X_replic = np.load("../data/sample/repliclust_dataset.npy")
y_replic = np.load("../data/sample/repliclust_labels.npy")


# Set plot style
rcParams['ps.fonttype'] = 42
plt.rcParams.update({
    "font.family": "serif",
    "font.serif": ["Times New Roman", "Times"],
    "text.usetex": False,
    "font.size": 10,
    "axes.titlesize": 12,
    "axes.labelsize": 10,
    "xtick.labelsize": 8,
    "ytick.labelsize": 8,
})

color_list = ['#000000', '#1b9e77', '#d95f02', '#7570b3',
              '#e7298a', '#66a61e', '#e6ab02', '#a6761d',
              '#666666', '#999999']
markers = ['o', 's', '^', 'v', '<', '>', 'P', '*', 'X', 'D']


def plot_clusters(ax, X, y, title):
    unique_labels = np.unique(y)
    for i, label in enumerate(unique_labels):
        idx = y == label
        ax.scatter(X[idx, 0], X[idx, 1],
                   color=color_list[i % len(color_list)],
                   marker=markers[i % len(markers)],
                   edgecolor='white',
                   linewidth=0.3,
                   s=25)
    ax.set_title(title, loc='left')
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.tick_params(axis='both', which='major')
    ax.grid(False)


# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(4, 5.5))

# Plot both
plot_clusters(ax1, X_comin, y_comin, '(a)')
plot_clusters(ax2, X_replic, y_replic, '(b)')

plt.tight_layout()
plt.subplots_adjust(hspace=0.3)

# Save final figure
fig.savefig("../results/fig1_scatter_plot.pdf", dpi=600, bbox_inches='tight', format='pdf', backend='cairo')
plt.show()
```
